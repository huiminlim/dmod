Create the symbols to be modulated using `numpy`, by generating an array of random `0`s and `1`s.

```python
# Fix the seed to a number to reproduce
np.random.seed(101)

# Create the symbols to modulate
symbols = np.random.default_rng().integers(0, 2, 100)

# Print the symbols 
print(symbols)
```
