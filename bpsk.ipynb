# BPSK

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

Create the symbols to be modulated using `numpy`, by generating an array of random `0`s and `1`s.

```python
# Fix the seed to a number to reproduce
np.random.seed(12888)

# Create the symbols to modulate
symbols = np.random.default_rng().integers(0, 2, 100)
time = np.arange(symbols.shape[0])

# Print the symbols 
print(symbols)
```

Plotting the symbols over time

```python
plt.ylim([-0.3, 1.3])
plt.yticks([0, 1])
plt.step(time, symbols, where='pre')
plt.show()
```
